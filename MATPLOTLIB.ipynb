```python
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
df = pd.DataFrame({
    "Ano": [2020,2021,2022,2023,2024,2025],
    "Vendas": [2000,150,800,180,500,1000]})

plt.plot(df["Ano"], df["Vendas"]) # .plot = gráfico linha
plt.title("Vendas Anuais") # Título do gráfico
plt.xlabel("Ano") # Se já tem title não precisa do xlabel
plt.ylabel("Vendas")
plt.grid(True)
plt.show() # Mostrar o gráfico

# 1º df é o eixo X (horizontal) e o 2º df é o eixo Y (vertical)
# É comum colocar os meses como números (1,2,3,4,5 - Jan, Fev, Mar, Abr, Mai)
# Ler o livro = Storytelling com Dados
# Grafico de Linha é para dado temporal
```

```python
df = pd.DataFrame({
    "Produto": ["A","B","C","D"],
    "Vendas": [500,700,300,900]})

df.sort_values(by="Vendas", ascending = False , inplace = True) # Ordenar o gráfico de barras

plt.figure(figsize=(10,7)) # Alterar o tamanho do gráfico
plt.barh(df["Produto"], df["Vendas"],color = "darkgreen") # .BAR = Gráfico de barras - COLOR = cor da barra - .BARH = Barra Horizontal
plt.title("Vendas por Produto") 
plt.xlabel("Produto") 
plt.ylabel("Vendas")
# COMANDO PARA SALVAR O GRÁFICO
plt.savefig("grafico_barras.png")
plt.show() 

# No de barra não necessita do grid
# Dentro do própio PowerBI da pra gerar o mesmo gráfico com esses comandos
# Jupyter Notebook só é recomendando para poucos dados
```

```python
df = {"Dias": ["Segunda","Terça","Quarta","Quinta","Sexta","Sábado","Domingo"],
      "Vendas": [150,200,300,250,400,500,450]}

dias = df["Dias"] # Melhor prática em gráficos usar variáveis para receber os dados
valores = df["Vendas"]

plt.plot(dias, valores, marker='o', linestyle='-', color='deeppink') # MARKER = marcador - LINESTYLE = estilo da linha
plt.title("Vendas por Dia")
plt.grid(True)
plt.show()

# No gráfico de linha não faz sentido ordenar do maior para o menor
```
