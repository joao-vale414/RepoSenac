```python
import random
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
dados = np.random.normal(loc=50, scale=10, size=100)
dados = [random.randint(1, 100) for i in dados]
dados.append(200)
```

```python
media = np.mean(dados)
mediana = np.median(dados)
distancia = (media - mediana) / mediana
desvio_padrao = np.std(dados)
variancia = np.var(dados)
q1 = np.percentile(dados, 25)
q2 = np.percentile(dados, 50)
q3 = np.percentile(dados, 75)
minimo = np.min(dados)
maximo = np.max(dados)
iqr = q3 - q1
cv = desvio_padrao / media # Coeficiente de variação
distancia_variancia = variancia / (media ** 2) # Coeficiente de Pearson
```

```python

fig, axs = plt.subplots(1, 2, figsize=(12, 6))

axs[0].boxplot(dados, vert=False,showmeans=True)
axs[0].set_title("Boxplot da Amostra")                                      

axs[1].axis("off")
axs[1].set_title("Estatísticas Descritivas")
axs[1].text(0.1, 0.9, f"Média: {media:.2f}")
axs[1].text(0.1, 0.8, f"Mediana: {mediana:.2f}")
axs[1].text(0.1, 0.7, f"Desvio Padrão: {desvio_padrao:.2f}")
axs[1].text(0.1, 0.6, f"Variância: {variancia:.2f}")
axs[1].text(0.1, 0.5, f"Q1: {q1:.2f}")
axs[1].text(0.1, 0.4, f"Q2: {q2:.2f}")
axs[1].text(0.1, 0.3, f"Q3: {q3:.2f}")
axs[1].text(0.1, 0.2, f"Mínimo: {minimo:.2f}")
axs[1].text(0.1, 0.1, f"IQR: {iqr:.2f}")
axs[1].text(0.1, 0.0, f"Distância: {(distancia):.4f}")
axs[1].text(0.1, -0.1, f"Coeficiente de variação: {(cv):.4f}")

plt.tight_layout()
plt.show()
```
