```python
idades = [12,25,14,30,15]

for i in idades:
  if i >= 18:
   print(i)
  else:
   print("Menor")

 # Fazendo dessa forma o if será implementado em cada valor dentro da lista
 # Caso fosse adicionado um else: print("menor") todos os valores menores que 18 seriam exibidos para o usuário com a palavra "menor"
```

```python
vazia = []

while True:
 try:
  nome = input("[Digite seu nome]: ")
  vazia.append(nome)
  sair = input("Deseja adicionar mais nome (s/n)?").upper()
  if sair == "N":
   break
 except:
  print("error")

print(vazia)
```
