MATPLOTLIB RESUMO

```python

import numpy as np
import matplotlib.pyplot as plt

# Criar uma figura e um eixo (opcional, mas útil para controle)
fig, ax = plt.subplots()

# 1. Plotar um gráfico simples de linha
x = [1, 2, 3, 4, 5]
y = [10, 8, 6, 4, 2]
plt.plot(x, y)  # linha conectando os pontos (x, y)

# 2. Adicionar título ao gráfico
plt.title("Título do Gráfico")

# 3. Definir rótulo do eixo X
plt.xlabel("Eixo X")

# 4. Definir rótulo do eixo Y
plt.ylabel("Eixo Y")

# 5. Mostrar a grade no gráfico
plt.grid(True)

# 6. Adicionar legenda (necessita de labels nos plots)
plt.plot(x, y, label="Dados")
plt.legend()  # mostra legenda com o label definido

# 7. Mostrar o gráfico na tela
plt.show()

# 8. Salvar o gráfico em um arquivo (PNG, JPG, SVG, PDF...)
plt.savefig("meu_grafico.png")

# 9. Plotar gráfico de barras simples
categorias = ['A', 'B', 'C']
valores = [5, 7, 3]
plt.bar(categorias, valores)

# 10. Plotar gráfico de barras empilhadas (stacked)
valores1 = [5, 7, 3]
valores2 = [2, 4, 6]
plt.bar(categorias, valores1, label='Grupo 1')
plt.bar(categorias, valores2, bottom=valores1, label='Grupo 2')  # empilhado em cima do grupo 1
plt.legend()

# 11. Alterar tamanho da figura
plt.figure(figsize=(10, 5))

# 12. Alterar cor da linha ou barra
plt.plot(x, y, color='red')  # linha vermelha
plt.bar(categorias, valores, color='green')  # barras verdes

# 13. Alterar estilo da linha
plt.plot(x, y, linestyle='--')  # linha tracejada

# 14. Alterar marcador nos pontos
plt.plot(x, y, marker='o')  # pontos circulares

# 15. Rotacionar labels do eixo X
plt.xticks(rotation=45)

# 16. Limitar o intervalo dos eixos
plt.xlim(0, 6)
plt.ylim(0, 12)
```

```python
import pandas as pd  # Importa a biblioteca pandas para manipulação de dados
import matplotlib.pyplot as plt  # Importa o matplotlib para criar gráficos

# Dados de exemplo: vendas por ano
vendas = pd.DataFrame({  # Cria um DataFrame com anos e valores de vendas
    "ano": [2018, 2019, 2020, 2021, 2022],  # Lista dos anos
    "valor": [15000, 18000, 17000, 22000, 21000]  # Valores de vendas para cada ano
})

# Agrupar e somar vendas por ano (caso tenha vários registros)
vendas_ano = vendas.groupby("ano")["valor"].sum().reset_index()  # Agrupa vendas por ano e soma os valores, criando um novo DataFrame

# Calcular vendas acumuladas
vendas_ano["valor_acumulado"] = vendas_ano["valor"].cumsum()  # Calcula o total acumulado de vendas até cada ano

# Calcular média de vendas
media_vendas = vendas_ano["valor"].mean()  # Calcula a média das vendas anuais

# Criar figura com 2 linhas e 2 colunas (4 blocos)
fig, axs = plt.subplots(2, 2, figsize=(12, 10))  # Cria uma figura com 4 subplots organizados em 2 linhas e 2 colunas

# Gráfico 1: Barras de vendas por ano
axs[0, 0].bar(vendas_ano["ano"].astype(str), vendas_ano["valor"], color="blue")  # Cria um gráfico de barras para vendas anuais
axs[0, 0].set_title("Vendas por Ano")  # Define o título do gráfico
axs[0, 0].set_xlabel("Ano")  # Define o rótulo do eixo X
axs[0, 0].set_ylabel("Valor das Vendas")  # Define o rótulo do eixo Y
axs[0, 0].grid(True, axis="y")  # Adiciona grade horizontal para facilitar a leitura

# Gráfico 2: Linha de vendas acumuladas
axs[0, 1].barh(vendas_ano["ano"].astype(str), vendas_ano["valor_acumulado"],color="darkgreen")  # Cria um gráfico de linha para vendas acumuladas
axs[0, 1].set_title("Vendas Acumuladas")  # Define o título do gráfico
axs[0, 1].set_xlabel("Ano")  # Define o rótulo do eixo X
axs[0, 1].set_ylabel("Valor Acumulado")  # Define o rótulo do eixo Y
axs[0, 1].grid(True, axis = "x")  # Adiciona grade ao gráfico

# Gráfico 3: Linha com média das vendas
axs[1, 0].plot(vendas_ano["ano"].astype(str), vendas_ano["valor"], marker="s", label="Vendas", color="purple")  # Plota as vendas anuais como pontos conectados por linha
axs[1, 0].axhline(media_vendas, color="red", linestyle="--", label=f"Média: {media_vendas:.2f}")  # Adiciona uma linha horizontal indicando a média das vendas
axs[1, 0].set_title("Vendas com Média")  # Define o título do gráfico
axs[1, 0].set_xlabel("Ano")  # Define o rótulo do eixo X
axs[1, 0].set_ylabel("Valor das Vendas")  # Define o rótulo do eixo Y
axs[1, 0].legend()  # Mostra a legenda para o gráfico
axs[1, 0].grid(True, axis = "y")  # Adiciona grade ao gráfico

# Bloco 4: Comentários/Observações 
axs[1, 1].axis("off")  # Desliga os eixos para o quarto bloco ficar só com texto

# Adiciona o texto de comentários no quarto bloco, alinhado à esquerda e centralizado verticalmente
axs[1, 1].text(0, 0.9, "Observações:", fontsize=14)  
axs[1, 1].text(0, 0.7, "- 2018 foi o pior ano!", fontsize=12)  
axs[1, 1].text(0, 0.6, "- As vendas aumentaram significativamente em 2021", fontsize=12)  
axs[1, 1].text(0, 0.5, "- 2022 tivemos uma queda", fontsize=12) 

# Ajusta o layout para evitar sobreposição dos gráficos e textos
fig.tight_layout()

# Define um título geral para a figura
fig.suptitle("Análise Completa de Vendas")  
fig.subplots_adjust(top=0.92)  # Ajusta espaço no topo para o título geral

# Exibe a figura com todos os gráficos e o bloco de comentários
plt.show()  
```
