```python
# Cardpio e Funes Principais do Cdigo 

cardapio = [
    {"N": 1, "Prato": "Lamen", "Descricao": "Macarro Japons No Caldo", "Valor": 25.0},
    {"N": 2, "Prato": "Yakisoba", "Descricao": "Macarro Com Legumes E Carne", "Valor": 30.0},
    {"N": 3, "Prato": "Guioza", "Descricao": "Pastelzinho Japons Recheado", "Valor": 18.0},
    {"N": 4, "Prato": "Sushi", "Descricao": "Bolinho De Arroz Com Peixe Cru", "Valor": 28.0},
    {"N": 5, "Prato": "Sashimi", "Descricao": "Fatias Finas De Peixe Cru", "Valor": 32.0},
    {"N": 6, "Prato": "Temaki", "Descricao": "Cone De Alga Com Recheio", "Valor": 26.0},
    {"N": 7, "Prato": "Onigiri", "Descricao": "Bolinhos De Arroz Enrolados Em Alga", "Valor": 15.0},
    {"N": 8, "Prato": "Okonomiyaki", "Descricao": "Panqueca Japonesa Salgada", "Valor": 27.0},
    {"N": 9, "Prato": "Takoyaki", "Descricao": "Bolinhos Recheados Com Polvo", "Valor": 22.0},
    {"N": 10, "Prato": "Katsudon", "Descricao": "Tigela De Arroz Com Porco Empanado", "Valor": 35.0},
    {"N": 11, "Prato": "Tonkatsu", "Descricao": "Fil De Porco Empanado Crocante", "Valor": 33.0},
    {"N": 12, "Prato": "Donburi", "Descricao": "Arroz Com Diversas Coberturas", "Valor": 29.0},
    {"N": 13, "Prato": "Ramen Miso", "Descricao": "Lamen Com Caldo De Miss", "Valor": 31.0},
    {"N": 14, "Prato": "Udon", "Descricao": "Macarro Grosso Servido No Caldo", "Valor": 24.0},
    {"N": 15, "Prato": "Tempur", "Descricao": "Legumes E Frutos Do Mar Empanados", "Valor": 34.0},
    {"N": 16, "Prato": "Karaage", "Descricao": "Frango Frito Japons Crocante", "Valor": 28.0},
    {"N": 17, "Prato": "Chahan", "Descricao": "Arroz Frito Estilo Japons", "Valor": 20.0},
    {"N": 18, "Prato": "Mochi", "Descricao": "Doce Japons De Arroz", "Valor": 12.0},
    {"N": 19, "Prato": "Dorayaki", "Descricao": "Panquecas Recheadas Com Doce De Feijo", "Valor": 14.0},
    {"N": 20, "Prato": "Matcha Latte", "Descricao": "Ch Verde Japons Com Leite", "Valor": 16.0},
]

pedidos = {}
proximo_id = 1

# 1. Funo para Exibir o Cardpio: 

def menu_cardapio():
 print("=====[CARDPIO]=====")
 for item in cardapio:
  print(f"{item['N']} - {item["Prato"]} ({item["Descricao"]}) - R$ {item["Valor"]:.2f}")
 
# 2. Funo para Registrar Pedidos: 

def registrar_pedido():
    global proximo_id
    mesa = input("Mesa: ")
    garcom = input("Garom: ")
    itens_pedido = []
    total = 0 

    while True:
     item_numero = int(input("Digite o N do item (0 para finalizar): "))
     if item_numero == 0:
      break 
     quantidade = int(input("Quantidade: "))

     item = next((i for i in cardapio if i["N"] == item_numero), None)

    if item:
      subtotal = item["Valor"] * quantidade
      itens_pedido.append({
        "Prato": item["Prato"],
        "Qtd": quantidade,
        "Valor": item["Valor"],
        "Subtotal": subtotal
    })
      total += subtotal
    else:
     print("No encontrado.")

    pedidos[proximo_id] = {
        "Mesa": mesa,
        "Garcom": garcom,
        "Itens": itens_pedido,
        "Total": total,
        "Status": "Aberto"
    }

    print(f"Pedido registrado com N {proximo_id}")
    proximo_id += 1

# 3. Funo Fechar Conta

def fechar():
    pedido_id = int(input("Nmero do pedido: "))
    
    if pedido_id not in pedidos:
        print("Pedido no encontrado.")
        return 
    pedido = pedidos[pedido_id]

    print("\n=====[CONTA]=====")
    print(f"Pedido: {pedido_id}")
    print(f"Mesa: {pedido["Mesa"]}")
    print(f"Garom: {pedido["Garcom"]}")
    print("\nItens:")

    for item in pedido["itens"]:
        print(f"{item["Prato"]} x{item["Qtd"]} - R$ {item["Subtotal"]:.2f}")
    
    print(f"\nTOTAL: R$ {pedido["Total"]:.2f}")

    pedido["Status"] = "Fechado"
   
```

```python
# Menu Oficial 

while True:
 print("\n=====[MENU]=====")
 print("1 - Cardpio")
 print("2 - Registrar pedido")
 print("3 - Fechar conta")
 print("0 - Sair")

 opcao = input("Escolha: ")
 if opcao == "1":
  menu_cardapio()
 elif opcao == "2":
  registrar_pedido()
 elif opcao == "3":
  fechar()
 elif opcao == "0":
  print("Encerrando...")
  break
 else:
  print("!!Selecione um opo Vlida!!")
```
